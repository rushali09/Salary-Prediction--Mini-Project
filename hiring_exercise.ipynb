```python
# In exercise folder (same level as this notebook on github) there is hiring.csv. This file contains hiring statics for a firm such as experience of candidate
# written test score and personal interview score. Based on these 3 factors, HR will decide the salary. Given this data, you need to build a machine learning
# model for HR department that can help them decide salaries for future candidates. Using this predict salaries for following candidates,
# 2 yr experience, 9 test score, 6 interview score
# 12 yr experience, 10 test score, 10 interview score
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import word2number as w2n

```

```python
hiring= pd.read_csv('hiring.csv', index_col= False)
```

```python
# Converting experience column values into number 
#word2number


!pip install word2number
```

```python
hiring['experience']
```

```python
from word2number import w2n

# Example usage
word = "twenty"
num = w2n.word_to_num(word)
print(num)  # This should print: 20
```

```python
hiring.experience= hiring.experience.apply(lambda x: w2n.word_to_num(x) if isinstance(x, str) else x)
```

```python
hiring.experience.median()
```

```python
hiring.experience= hiring.experience.fillna(hiring.experience.median())
```

```python
hiring
```

```python
hiring = hiring.rename(columns={'test_score(out of 10)': 'test_score'})
hiring = hiring.rename(columns={'interview_score(out of 10)': 'interview_score'})
hiring = hiring.rename(columns={'salary($)': 'salary'})
```

```python
hiring
```

```python
hiring.test_score= hiring.test_score.fillna(hiring.test_score.median())
```

```python
hiring
```

```python
# Creating object
reg= linear_model.LinearRegression()
```

```python
reg.fit(hiring[['experience', 'test_score', 'interview_score']], hiring.salary)
```

```python
# 2 yr experience, 9 test score, 6 interview score
reg.predict([[2,9,6]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# 12 yr experience, 10 test score, 10 interview score
reg.predict([[12, 10, 10]])
```
